```python
import gensim#For word2vec, etc
from gensim.models import ldaseqmodel
import matplotlib.pyplot as plt
import numpy as np
```

```python
dtm_left = ldaseqmodel.LdaSeqModel.load("data_ldaseq_5yr_left")
dtm_right = ldaseqmodel.LdaSeqModel.load("data_ldaseq_5yr_right")
```

```python
dtm_left.dtm_coherence(4)
```

```python
for i in range(5):
    print(dtm_left.print_topic_times(topic=1)[i][0][1])
```

```python
for i in range(5):
    print(dtm_left.print_topic_times(topic=1)[i][10][1])
```

```python
fig, axes = plt.subplots(10, 2, figsize = (14, 18))

ax = axes.ravel()
for i in range(20):
    evolution = []
    for j in range(5):
        evolution.append(dtm_left.print_topic_times(topic=0)[j][i][1])
    ax[i].plot(evolution)
    ax[i].set_xticks(np.arange(1,5))
    ax[i].set_title(dtm_left.print_topic_times(topic=0)[j][i][0])
plt.tight_layout(pad=1)
        
```
